```python
#Import Statements
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read Dataset 1
df1 = pd.read_csv(r'I:\Human Computer Interaction\Statistical Analysis\Statistical Analysis\dataset1.csv')
```

```python
# Read Dataset 2
df2 = pd.read_csv(r'I:\Human Computer Interaction\Statistical Analysis\Statistical Analysis\dataset2.csv')
```

```python
# Boxplot for Task 1 - Menu vs Time
plt.figure(figsize=(10, 10))
ax=sns.boxplot(x=df1["menu"],y=df1["time"], palette='twilight_shifted')
ax = sns.swarmplot(x=df1["menu"],y=df1["time"], color="seagreen")
```

```python
# Boxplot for Task 2 - Menu vs Time
plt.figure(figsize=(10, 10))
ax=sns.boxplot(x=df2["menu"],y=df2["time"], palette='Pastel1')
ax = sns.swarmplot(x=df2["menu"],y=df2["time"], color="blue")
```

```python
# Boxplot for Task 2 - Menu vs Error
plt.figure(figsize=(10, 10))
ax=sns.boxplot(x=df2["menu"],y=df2["error"], palette='Purples')
ax = sns.swarmplot(x=df2["menu"],y=df2["error"], color="red")
```

```python
# Barplot for Task 1 - Mean of Time for each type of Menu
times=pd.DataFrame()
times['Mean of Time'] = df1.groupby(['menu']).mean()['time']
times['menu'] = times.index
group_top = times.sort_values(by='Mean of Time')

plt.figure(figsize=(10, 5))
sns.set(color_codes=True)
sns.set(font_scale = 1.5)
ax = sns.barplot(x="menu", y="Mean of Time",palette='GnBu_r', data=group_top)
font_size= {'size': 'medium'}
ax.set_title("Mean of Time for each type of Menu", **font_size)
xt = plt.xticks(rotation=65)
```

```python
# Barplot for Task 2 - Mean of Time for each type of Menu
times=pd.DataFrame()
times['Mean of Time'] = df2.groupby(['menu']).mean()['time']
times['menu'] = times.index
group_top = times.sort_values(by='Mean of Time')

plt.figure(figsize=(10, 5))
sns.set(color_codes=True)
sns.set(font_scale = 1.5)
ax = sns.barplot(x="menu", y="Mean of Time",palette='ocean_r', data=group_top)
font_size= {'size': 'medium'}
ax.set_title("Mean of Time for each type of Menu", **font_size)
xt = plt.xticks(rotation=65)
```

```python
# Barplot for Task 2 - Mean of Error for each type of Menu
times=pd.DataFrame()
times['Mean of Error'] = df2.groupby(['menu']).mean()['error']
times['menu'] = times.index
group_top = times.sort_values(by='Mean of Error')

plt.figure(figsize=(10, 5))
sns.set(color_codes=True)
sns.set(font_scale = 1.5)
ax = sns.barplot(x="menu", y="Mean of Error",palette='ocean', data=group_top)
font_size= {'size': 'medium'}
ax.set_title("Mean of Time for each type of Menu", **font_size)
xt = plt.xticks(rotation=65)
```
